```python
from wdgm_library import *
```

```python
book_img = cv2.imread('book.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(book_img, cmap='gray')
```

```python
book_thresh = cv2.adaptiveThreshold(
    book_img,
    maxValue=255,
    adaptiveMethod=cv2.ADAPTIVE_THRESH_MEAN_C,
    thresholdType=cv2.THRESH_BINARY,
    blockSize=23,
    C=1.9
)
```

```python
plt.imshow(book_thresh, cmap='gray')
```

```python
book_edges = cv2.Canny(
      book_thresh,
      5,  # prog histerezy 1
      4,  # prog histerezy 2
      3  # wielkoscc filtra sobela
)
```

```python
plt.imshow(book_edges, cmap='gray')
```

```python
lines = cv2.HoughLinesP(
    book_edges,
    1,
    np.pi / 180,
    30
)
```

```python
result_lines_img = cv2.cvtColor(book_img, cv2.COLOR_GRAY2RGB)
```

```python
for line in lines:
  x0, y0, x1, y1 = line[0]
  cv2.line(result_lines_img, (x0, y0), (x1, y1), (0, 0, 255), 2)
```

```python
plt.imshow(result_lines_img)
```
